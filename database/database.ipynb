```python
# importing all packages
#importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import yfinance as yf
import datetime
import time
import requests
import io
import os
```

```python
#start and end times for our stock dataframe
start = datetime.datetime(2022,2,1)
end = datetime.datetime(2022,4,18)
```

```python
#instructions to convert stock market data into dataframe https://towardsdatascience.com/downloading-historical-stock-prices-in-python-93f85f059c1f
url="https://pkgstore.datahub.io/core/nasdaq-listings/nasdaq-listed_csv/data/7665719fb51081ba0bd834fde71ce822/nasdaq-listed_csv.csv"
s = requests.get(url).content
companies = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
#making a symbols a list to read stock names
Symbols = companies['Symbol'].tolist()
```

```python
# create empty dataframe
stock_final = pd.DataFrame()
```

```python

# iterate over each symbol
for i in Symbols:  
    
    try:
        # download the stock price 
        stock = []
        stock = yf.download(i,start=start, end=end, progress=False)
        
        # append the individual stock prices 
        if len(stock) == 0:
            None
        else:
            stock['Name']=i
            stock_final = stock_final.append(stock,sort=False)
    except Exception:
        None
```

```python
#get rid of volume and adj close dont need them
stock_final2=stock_final[['Open','High','Low','Close','Name']]
print(stock_final2.tail())

```

```python
import sqlite3
```

```python
conn = sqlite3.connect('stock_database.sqlite')
c = conn.cursor()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS stocks (Open number,High number,Low number ,Close number,Name text)')
conn.commit()
```

```python
stock_final2.to_sql('stocks', conn, if_exists='replace', index = True)
```

```python
print(pd.read_sql_query("SELECT * FROM stocks", conn))
```
